```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
# from selenium.webdriver.common.keys import Keys
import time
# from datetime import datetime, timedelta
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
# from selenium.webdriver import ChromeOptions

import logging

from webdriver_manager.chrome import ChromeDriverManager

# options = ChromeOptions()
chrome_options = Options() 
# chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
# chrome_options.add_experimental_option('useAutomationExtension', False)
user_data_dir = "C:/Users/Siddhant/AppData/Local/Google/Chrome/User Data"
profile_dir = "Default" 
# chrome_options.add_argument(f"user-data-dir={user_data_dir}")  # Path to user data
chrome_options.add_argument(f"profile-directory={profile_dir}")
chrome_driver_path = "C:\\Users\\Siddhant\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe"
driver = webdriver.Chrome(options=chrome_options)

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s', filemode='w')
logging.info("Starting WebDriver session")

```

```python
try:
    # Step 1: Navigate to the website
    driver.get("https://www.goindigo.in/")
    logging.info("Navigated to the website")

    # Verify landing on the correct page
    assert "goindigo.in" in driver.current_url
    time.sleep(5)
    # Wait for the page to load
    wait = WebDriverWait(driver, 10)
    
    
    try:
        # Wait for the pop-up's close button and click it
        close_ad = wait.until(EC.element_to_be_clickable((By.XPATH, "//span[@class='flight-close']")))
        close_ad.click()
        logging.info("Advertisement closed")
    except:
        print("No advertisement found.")
    time.sleep(5)
    # Step 2: Click on "Book" option
    book_dropdown = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(@class,'headerv2__navbar-item') and contains(text(), 'Book')]")))
    book_dropdown.click()
    logging.info("Clicked on 'Book' button")
    time.sleep(5)
    
    # Step 3: Click on "Flight" option
    # wait.until(EC.presence_of_element_located((By.XPATH, "//title[contains(text(), 'Flight Booking')]")))
    flight_option = wait.until(EC.element_to_be_clickable((By.XPATH, "//a[@aria-label='CTA to Book Flights']")))
    flight_option.click()
    logging.info("Clicked on 'Flights' button")
    time.sleep(5)
    
    
except:
    # Close the browser after completion
    logging.error("An error occcured : {e}")
    time.sleep(3)
    driver.quit()
```

```python
import dateutil.relativedelta 

try:
    # Step 4: Click on the pax option and increase adult count to 2
    time.sleep(5)
    try:
        pax_option = wait.until(EC.element_to_be_clickable((By.XPATH, "//*[@id='bar']/div[2]/div[1]")))
        pax_option.click()
        logging.info("Clicked on '1 pax' button")
    except:
        logging.error("pax option not found")   
     
    time.sleep(5)
            
    
    try:
        pax_dropdown =wait.until(
            EC.presence_of_element_located((By.CLASS_NAME, "pax-dropdown__wrapper"))
        )
    except:
        logging.error("pax dropdown not found")       
    time.sleep(5)
    

    # Locate the "Add Adult" button inside the pax dropdown

    try:
        add_adult_button = pax_dropdown.find_element(By.CSS_SELECTOR, ".stepper-input__btn--plus")
        # Click the "Add Adult" button
        add_adult_button.click()
        logging.info("Increased adults to 2")
    except:
        logging.error("add adult button not found")
            
    time.sleep(5)
    
    


    
    done_button = wait.until(
        EC.element_to_be_clickable((By.CLASS_NAME, "pax-dropdown__action__done"))
    )

    # Click the "Done" button
    done_button.click()
    logging.info("Clicked on 'Done' button")
    time.sleep(5)
    
    # Step 5: Enter "Hyderabad" in the From option and select from dropdown
    from_field = driver.find_element(By.XPATH, "//*[@id='container-b9910c1891']/div/div[3]/div/div/div/section/div/div[3]/div[1]/div[1]/input")
    from_field.click()
    from_field.clear()
    from_field.send_keys("Hyderabad")
    time.sleep(1)  # Wait for the dropdown
    dropdown_option = wait.until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="container-b9910c1891"]/div/div[3]/div/div/div/section/div/div[3]/div[1]/div[2]/div[3]/div[1]/div[2]'))
    )
    dropdown_option.click()
    logging.info("Selected Hyderabad in 'From' field")
    time.sleep(5)
    
    
    to_field = driver.find_element(By.XPATH, '//*[@id="container-b9910c1891"]/div/div[3]/div/div/div/section/div/div[3]/div[2]/div[1]/input')
    to_field.click()
    to_field.clear()
    to_field.send_keys("Delhi")
    time.sleep(1)  # Wait for the dropdown
    dropdown_option = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="container-b9910c1891"]/div/div[3]/div/div/div/section/div/div[3]/div[2]/div[2]/div[3]/div[1]/div[2]'))
    )
    dropdown_option.click()
    logging.info("Selected Delhi in 'To' field")
    time.sleep(5)

    # Step 7: Select a travel date 1 month from today
    travel_dates_input = driver.find_element(By.XPATH, '//*[@id="travel-dates-container"]/div[1]/div/div/input')
    js_selector = "#travel-dates-container > div.custom-form-control.input-text-field.from-date > div.react-datepicker-wrapper > div > input"
    # Calculate the date one month from now
    import datetime
    import dateutil
    today = datetime.date.today()
    print(today)
    next_month = today + dateutil.relativedelta.relativedelta(months=1)
    date_one_month_from_now = next_month.strftime("%d %b %Y")  # Format as "26 Oct 2024"
    print(date_one_month_from_now)
    time.sleep(2)

    
    driver.execute_script("document.querySelector(arguments[0]).value = '';", js_selector)  # Clear the input field
    time.sleep(2)
    driver.execute_script("document.querySelector(arguments[0]).value = arguments[1];", js_selector, date_one_month_from_now)  # Set the new date  
    time.sleep(2)
    logging.info("Selected date one month from today")
    
    search_flight_button = wait.until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="container-b9910c1891"]/div/div[3]/div/div/div/section/div[1]/div[3]/button'))
    )
    driver.execute_script("argument[0].click()", search_flight_button)
    logging.info("Clicked on search flight button")

    
except: 
    logging.error("An error occured : {e}")
    time.sleep(3)
    driver.quit()
```
